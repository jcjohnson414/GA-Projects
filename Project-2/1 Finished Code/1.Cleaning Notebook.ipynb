### Created a function that would output a readable version of all of the unique values and value counts for each column.

```python
def columns_values(dataframe):
    for n in dataframe.columns:                               # cycle through the columns in dataframe
        column = n  
        value_counts = (dataframe[column].value_counts())     # run value counts for each column
        unique_values = (dataframe[column].unique())          # run unique values for each column 
        len_counts = len(value_counts)                        # how many items are counted in value counts
        len_unique = len(unique_values)                       # how many items are counted in unique values
        if (len(value_counts)< 100):                          # ignore columns where counts over 100
                print(f"{column} has length of {len_unique}") # print how many unique values there are
                print(unique_values)                          # print unique values for column 
                print("="*30)                                  
                print(f"{column} has length of {len_counts}") # print how many values are counted
                print(value_counts)                           # print unique values counted
                print("="*60)
```

### Here I created two functions to completed data cleaning for the training and test sets, this was updated throughout the EDA process.

```python
# Credit to Haley Taft for showing me how this could work.
def clean_train(dataframe, inplace=True):
    ## BELOW HERE IS WHERE I SET MY MAPPING DICTIONARIES TO REPLACE VALUES IN THE DATAFRAME ##
    replace_values = {'Reg': 4, 'IR1': 3, 'IR2': 2, 'IR3': 1, 
     'AllPub': 4, 'NoSewr': 3, 'NoSeWa': 2, 'ELO': 1,
     'Gtl': 3, 'Mod': 2, 'Sev': 1,
     'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1,
     'Gd': 4, 'Av': 3, 'Mn': 2, 'No':1,
     'GLQ': 6, 'ALQ': 5, 'BLQ': 4, 'Rec': 3, 'LwQ': 2, 'Unf': 1,
     'N': 0, 'Y': 1,
     'SBrkr': 5, 'FuseA': 4, 'FuseF': 3, 'FuseP': 2, 'Mix': 1,
     'Typ': 8, 'Min1': 7, 'Min2': 6, 'Mod': 5, 'Maj1': 4, 'Maj2': 3, 'Sev': 2, 'Sal': 1,
     'Fin': 3, 'RFn': 2, 'Unf': 1,
     'Y': 3, 'P': 2, 'N': 1,
     'GdPrv': 4, 'MnPrv': 3, 'GdWo': 2, 'MnWw': 1, 'NA' : 0
     }
    replace_values_na = {np.nan: "NA"}
    replace_values_nan = {np.nan: 0}
    
    ## HERE IS WHERE I USED THE MAPPING DICTIONARIES TO REPLACE THE VALUES
    dataframe['Bsmt Qual'] = dataframe['Bsmt Qual'].replace(replace_values_nan)
    dataframe['Bsmt Cond'] = dataframe['Bsmt Cond'].replace(replace_values_nan)
    dataframe['Bsmt Exposure'] = dataframe['Bsmt Exposure'].replace(replace_values_nan)
    dataframe['BsmtFin Type 1'] = dataframe['BsmtFin Type 1'].replace(replace_values_nan)
    dataframe['BsmtFin Type 2'] = dataframe['BsmtFin Type 2'].replace(replace_values_nan)
    dataframe['Bsmt Full Bath'] = dataframe['Bsmt Full Bath'].replace(replace_values_nan)
    dataframe['Bsmt Half Bath'] = dataframe['Bsmt Half Bath'].replace(replace_values_nan)
    dataframe['Fireplace Qu'] = dataframe['Fireplace Qu'].replace(replace_values_nan)
    dataframe['Garage Finish'] = dataframe['Garage Finish'].replace(replace_values_nan)
    dataframe['Garage Qual'] = dataframe['Garage Qual'].replace(replace_values_nan)
    dataframe['Garage Cond'] = dataframe['Garage Cond'].replace(replace_values_nan)
    dataframe['Fireplace Qu'] = dataframe['Fireplace Qu'].replace(replace_values_nan)
    dataframe['Garage Cars'] = dataframe['Garage Cars'].replace(replace_values_nan)
    dataframe['Garage Area'] = dataframe['Garage Area'].replace(replace_values_nan)
    dataframe['Pool QC'] = dataframe['Pool QC'].replace(replace_values_nan)
    dataframe['Fence'] = dataframe['Fence'].replace(replace_values_nan)
    dataframe['Lot Frontage'] = dataframe['Lot Frontage'].replace(replace_values_nan)
    dataframe['Alley'] = dataframe['Alley'].replace(replace_values_na)
    dataframe['Mas Vnr Type'] = dataframe['Mas Vnr Type'].replace(replace_values_na)
    dataframe['Mas Vnr Area'] = dataframe['Mas Vnr Area'].replace(replace_values_nan)
    dataframe['Mas Vnr Area'] = dataframe['Mas Vnr Area'].replace(replace_values_nan)
    dataframe['BsmtFin SF 1'] = dataframe['BsmtFin SF 1'].replace(replace_values_nan)
    dataframe['BsmtFin SF 2'] = dataframe['BsmtFin SF 2'].replace(replace_values_nan)
    dataframe['Bsmt Unf SF'] = dataframe['Bsmt Unf SF'].replace(replace_values_nan)
    dataframe['Total Bsmt SF'] = dataframe['Total Bsmt SF'].replace(replace_values_nan)
    dataframe['Garage Type'] = dataframe['Garage Type'].replace(replace_values_na)
    dataframe['Garage Yr Blt'] = dataframe['Garage Yr Blt'].replace({np.nan: 1980})
    dataframe['Misc Feature'] = dataframe['Misc Feature'].replace(replace_values_na)
    dataframe['Electrical'] = dataframe['Electrical'].replace(replace_values_nan)
    dataframe = dataframe.replace(replace_values)
    
    ## AFTER EDA I DETERMINED THAT I WANTED TO REMOVE THE OUTLIERS BELOW BASED ON SQUARE FOOTAGE
    
    dataframe = dataframe[(dataframe['Gr Liv Area'] < 4000)]
    dataframe = dataframe[(dataframe['Lot Area'] < 40000)]
    
    ## AFTER EDA I DECIDED TO CREATE THIS FEATURE COLUMN
    dataframe['Exter Qual and Gr Liv Area'] = dataframe['Exter Qual'] * dataframe['Gr Liv Area']
   
    return dataframe
```
